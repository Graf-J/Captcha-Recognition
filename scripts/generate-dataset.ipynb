```python
from tqdm import tqdm
from pathlib import Path
from captcha.image import ImageCaptcha
from PIL import Image
import string
import random
```

```python
vocab = string.ascii_lowercase + string.ascii_uppercase + string.digits
output_dir = Path("/nfs/home/tpz8688/Captcha-Recognition/data/generated")
generation_plan = {
    1: 1000,
    2: 2000,
    3: 10000,
    4: 20000,
    5: 40000,
    6: 40000,
    7: 40000,
    8: 40000
}

total_images = sum(generation_plan.values())
print(f"Total images to generate: {total_images}")
```

```python
def generate_image(input_str):
    """Generates a captcha image variable from a string."""
    generator = ImageCaptcha(fonts=["/nfs/home/tpz8688/Captcha-Recognition/data/Nunito.ttf"])
    image_data = generator.generate(input_str)
    return Image.open(image_data)
```

```python
# Using a single global progress bar for all stages
with tqdm(total=total_images, desc="Total Progress") as pbar:
    for char_length, count in generation_plan.items():
        for i in range(count):
            # 1. Select random string of specific length
            random_str = ''.join(random.choices(vocab, k=char_length))
            
            # 2. Generate and save
            try:
                img = generate_image(random_str)
                # Filename includes length and index to prevent any collisions
                file_path = output_dir / f"{random_str}_len{char_length}_{i}.png"
                img.save(file_path)
            except Exception as e:
                print(f"Error generating '{random_str}': {e}")
            
            pbar.update(1)

print(f"Generation complete. Saved to {output_dir}")
```
